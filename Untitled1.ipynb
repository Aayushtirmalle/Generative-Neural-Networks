```python
import numpy as np
from sklearn.datasets import make_moons
from sklearn.metrics.pairwise import euclidean_distances
import matplotlib.pyplot as plt

# Generate 2D training data using make_moons
X, _ = make_moons(n_samples=100, noise=0.05, random_state=42)

# Define a function to calculate MMD with squared exponential kernel
def mmd_squared_exponential(X, Y, gamma):
    XY = euclidean_distances(X, Y)
    XX = euclidean_distances(X, X)
    YY = euclidean_distances(Y, Y)

    mmd = np.mean(np.exp(-gamma * XX)) + np.mean(np.exp(-gamma * YY)) - 2 * np.mean(np.exp(-gamma * XY))
    return mmd

# Define a function to create a 2D histogram model
def create_2d_histogram(X, bin_size):
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1

    x_bins = np.linspace(x_min, x_max, bin_size)
    y_bins = np.linspace(y_min, y_max, bin_size)

    histogram, _, _ = np.histogram2d(X[:, 0], X[:, 1], bins=(x_bins, y_bins))
    histogram /= np.sum(histogram)

    return histogram

# Create a 2D histogram model with a specific bin size
bin_size = 10
histogram_model = create_2d_histogram(X, bin_size)

# Generate test data
test_data, _ = make_moons(n_samples=100, noise=0.05, random_state=123)

# Calculate MMD using squared exponential kernel
gamma = 1.0  # Adjust the value of gamma for the squared exponential kernel
mmd = mmd_squared_exponential(X, test_data, gamma)

# Define x_min, x_max, y_min, y_max
x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1

# Visualize the histogram model and test data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(histogram_model, cmap='viridis', origin='lower', extent=[x_min, x_max, y_min, y_max])
plt.title("2D Histogram Model")

plt.subplot(1, 2, 2)
plt.scatter(test_data[:, 0], test_data[:, 1], c='b', marker='o', label='Test Data')
plt.title("Test Data")
plt.legend()

plt.suptitle(f"MMD with Squared Exponential Kernel: {mmd:.4f}")
plt.show()
```

```python
# Generate data from the 2D histogram model
def generate_data_from_histogram(histogram_model, num_samples):
    x_min, x_max, y_min, y_max = -2, 3, -1, 2  # Define the range of your data
    x_values = np.linspace(x_min, x_max, histogram_model.shape[1])
    y_values = np.linspace(y_min, y_max, histogram_model.shape[0])
    x_mesh, y_mesh = np.meshgrid(x_values, y_values)

    flattened_hist = histogram_model.flatten()
    flattened_hist /= flattened_hist.sum()  # Normalize the histogram

    # Sample from the flattened histogram
    data_indices = np.random.choice(len(flattened_hist), size=num_samples, p=flattened_hist)
    
    # Create the dataset based on the sampled indices
    generated_data = np.column_stack((x_mesh.flatten()[data_indices], y_mesh.flatten()[data_indices]))
    
    return generated_data


histogram_data = generate_data_from_histogram(histogram_model, num_samples=1000)

# Visualize the numerical values of the learned density
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(histogram_model, cmap='viridis', origin='lower', extent=[x_min, x_max, y_min, y_max])
plt.title("2D Histogram Model")

# Visualize the generated dataset
plt.subplot(1, 2, 2)
plt.scatter(histogram_data[:, 0], histogram_data[:, 1], c='b', marker='o', label='Generated Data')
plt.title("Generated Data from 2D Histogram Model")
plt.legend()

plt.suptitle("Model Visualization - 2D Histogram")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

# Generate 2D training data using make_moons
X, _ = make_moons(n_samples=100, noise=0.05, random_state=42)

# Define a function to calculate MMD with squared exponential kernel
def mmd_squared_exponential(X, Y, gamma):
    XX = np.dot(X, X.T)
    XY = np.dot(X, Y.T)
    YY = np.dot(Y, Y.T)

    mmd = np.mean(np.exp(-gamma * (XX - 2 * XY + YY)))
    return mmd

# Define a custom Kernel Density Estimator (KDE) with a squared exponential kernel
def custom_kde_squared_exponential(X, test_data, bandwidth):
    n_samples = X.shape[0]
    n_test_samples = test_data.shape[0]
    kde_scores = np.zeros(n_test_samples)

    for i in range(n_test_samples):
        distances = np.sum((X - test_data[i]) ** 2, axis=1)
        kde_scores[i] = np.sum(np.exp(-0.5 * distances / (bandwidth**2))) / (n_samples * (np.sqrt(2 * np.pi) * bandwidth)**2)

    return kde_scores

# Create a custom KDE with a specific bandwidth
bandwidth = 0.1  # Adjust the bandwidth for the squared exponential kernel

# Generate test data
test_data, _ = make_moons(n_samples=100, noise=0.05, random_state=123)

# Calculate MMD using squared exponential kernel
gamma = 1.0  # Adjust the value of gamma for the squared exponential kernel
mmd = mmd_squared_exponential(X, test_data, gamma)

# Visualize the custom KDE model and test data
x, y = np.mgrid[-2:3:0.01, -1:2:0.01]
pos = np.dstack((x, y))
kde_scores = custom_kde_squared_exponential(X, pos.reshape(-1, 2), bandwidth)
kde_scores = kde_scores.reshape(x.shape)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.contourf(x, y, kde_scores, cmap='viridis')
plt.title("Custom Kernel Density Estimator (KDE) with Squared Exponential Kernel")

plt.subplot(1, 2, 2)
plt.scatter(test_data[:, 0], test_data[:, 1], c='b', marker='o', label='Test Data')
plt.title("Test Data")
plt.legend()

plt.suptitle(f"MMD with Squared Exponential Kernel: {mmd:.4f}")
plt.show()
```
