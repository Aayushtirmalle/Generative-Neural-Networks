```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

# Generate the dataset
X, _ = make_moons(n_samples=1000, noise=0.05, random_state=42)
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

# 2D Histogram Model
def create_histogram(X, bins):
    H, xedges, yedges = np.histogram2d(X[:, 0], X[:, 1], bins=bins)
    return H, xedges, yedges

# Calculate MMD1 with squared exponential kernel
def mmd1_sq_exp(X, Y, sigma):
    XX = np.dot(X, X.T)
    YY = np.dot(Y, Y.T)
    XY = np.dot(X, Y.T)
    X_sq_norms = np.sum(X * X, axis=1, keepdims=True)
    Y_sq_norms = np.sum(Y * Y, axis=1, keepdims=True)
    
    # MMD1 with squared exponential kernel
    mmd = np.mean(np.exp(-0.5 / sigma**2) * (X_sq_norms - 2 * XY + Y_sq_norms))
    return mmd

# Example usage:
histogram, _, _ = create_histogram(X_train, bins=20)
histogram_test, _, _ = create_histogram(X_test, bins=20)
mmd_value = mmd1_sq_exp(histogram, histogram_test, sigma=1.0)
print("MMD Value:", mmd_value)

# Visualization (You will need to adapt this part for different models)
plt.imshow(histogram.T, origin='lower', extent=[xedges.min(), xedges.max(), yedges.min(), yedges.max()], cmap='Blues')
plt.colorbar()
plt.show()
```

```python
# Single Gaussian Model
def fit_single_gaussian(X):
    mean = np.mean(X, axis=0)
    cov = np.cov(X, rowvar=False)
    return mean, cov

def evaluate_single_gaussian(x, mean, cov):
    # Calculate the likelihood of a data point 'x' under the Gaussian distribution
    diff = x - mean
    exponent = -0.5 * np.dot(diff, np.dot(np.linalg.inv(cov), diff))
    return np.exp(exponent) / (2 * np.pi) * np.sqrt(np.linalg.det(cov))

# Fit the Single Gaussian model
mean, cov = fit_single_gaussian(X_train)

# Evaluate the Single Gaussian model on the test set
likelihoods = [evaluate_single_gaussian(x, mean, cov) for x in X_test]

# Visualization (You will need to adapt this part for different models)
plt.scatter(X_test[:, 0], X_test[:, 1], c=likelihoods, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
class GMM:
    def __init__(self, n_components, n_iterations=100, random_state=None):
        self.n_components = n_components
        self.n_iterations = n_iterations
        self.random_state = random_state
        self.means = None
        self.covs = None
        self.weights = None

    def initialize_parameters(self, X):
        # Initialize means, covariances, and weights randomly
        np.random.seed(self.random_state)
        n_samples, _ = X.shape
        random_samples = np.random.choice(n_samples, size=self.n_components, replace=False)
        self.means = X[random_samples]
        self.covs = [np.eye(X.shape[1]) for _ in range(self.n_components)]
        self.weights = np.ones(self.n_components) / self.n_components

    def calculate_likelihood(self, X):
        # Calculate the likelihood of data points given the current GMM parameters
        likelihood = np.zeros((X.shape[0], self.n_components))
        for k in range(self.n_components):
            likelihood[:, k] = self.weights[k] * self.multivariate_normal(X, self.means[k], self.covs[k])
        return likelihood

    def multivariate_normal(self, X, mean, cov):
        det_cov = np.linalg.det(cov)
        inv_cov = np.linalg.inv(cov)
        exponent = np.einsum('ij,ij->i', X - mean, np.dot(inv_cov, (X - mean).T).T)
        return np.exp(-0.5 * exponent) / np.sqrt((2 * np.pi) ** X.shape[1] * det_cov)

    def fit(self, X):
        self.initialize_parameters(X)
        for _ in range(self.n_iterations):
            likelihood = self.calculate_likelihood(X)
            responsibilities = likelihood / likelihood.sum(axis=1, keepdims=True)
            N_k = responsibilities.sum(axis=0)

            for k in range(self.n_components):
                self.means[k] = (1 / N_k[k]) * np.sum(responsibilities[:, k][:, np.newaxis] * X, axis=0)
                self.covs[k] = (1 / N_k[k]) * np.dot((responsibilities[:, k][:, np.newaxis] * (X - self.means[k])).T, (X - self.means[k]))
                self.weights[k] = N_k[k] / X.shape[0]

    def predict(self, X):
        likelihood = self.calculate_likelihood(X)
        return np.argmax(likelihood, axis=1)

# Fit the Gaussian Mixture Model
n_components = 2
gmm = GMM(n_components=n_components, n_iterations=100, random_state=42)
gmm.fit(X_train)

# Predict clusters for the test set
cluster_predictions = gmm.predict(X_test)

# Visualization
plt.scatter(X_test[:, 0], X_test[:, 1], c=cluster_predictions, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
# KDE with Squared Exponential Kernel
class KDE:
    def __init__(self, bandwidth):
        self.bandwidth = bandwidth

    def fit(self, X):
        self.X = X

    def evaluate(self, x):
        n_samples = self.X.shape[0]
        # Calculate the KDE using a squared exponential kernel
        kernel_vals = np.exp(-0.5 * ((x - self.X) / self.bandwidth) ** 2).sum(axis=0)
        kde_value = (1 / (n_samples * self.bandwidth * np.sqrt(2 * np.pi))) * kernel_vals
        return kde_value

# Fit the KDE model
bandwidth = 0.1
kde = KDE(bandwidth)
kde.fit(X_train)

# Evaluate the KDE on the grid for visualization
x, y = np.meshgrid(np.linspace(X_test[:, 0].min(), X_test[:, 0].max(), 100), np.linspace(X_test[:, 1].min(), X_test[:, 1].max(), 100))
grid_points = np.c_[x.ravel(), y.ravel()]
kde_values = np.array([kde.evaluate(point) for point in grid_points])

# Reshape the KDE values to match the grid
kde_values = kde_values.reshape(x.shape)

# Visualization with contour plot
plt.contourf(x, y, kde_values, cmap='viridis', levels=100)
plt.colorbar()
plt.scatter(X_test[:, 0], X_test[:, 1], color='blue', marker='o', s=10, label='Test Data')
plt.legend()
plt.show()
```
